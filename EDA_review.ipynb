```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("yelp_dataset_csv/yelp_academic_dataset_review.csv")
df.head()
```

```python
df_business = pd.read_csv("yelp_dataset_csv/yelp_academic_dataset_business.csv")
```

```python
df_business_philadelphia = df_business[(df_business["state"] == "PA") &
                                       (df_business["categories"].str.contains("Restaurants"))]
#clean the business ids
df_business_philadelphia_rest["business_id"] = (
    df_business_philadelphia_rest["business_id"]
    .str.replace(r"^\-+", "", regex=True)
)
df['business_id'] = (df['business_id'].str.replace(r"^\-+", "", regex=True))

df_business_philadelphia_review = pd.merge(df_business_philadelphia_rest, df, on="business_id")


```

```python
df_business_philadelphia_rest.count()
```

```python
df_business_philadelphia_review.groupby("business_id").size().reset_index(name="count")
```

```python
#validate

review_counts_actual = (
    df_business_philadelphia_review
    .groupby("business_id")
    .size()
    .reset_index(name="actual_review_count")
)

df_validate_review_count = pd.merge(
    df_business_philadelphia_rest[["business_id", "review_count"]],
    review_counts_actual,
    on="business_id",
    how="left"
)

df_validate_review_count["valid"] = (
    df_validate_review_count["review_count"] == df_validate_review_count["actual_review_count"]
)

```

```python
df_invalid = df_validate_review_count[df_validate_review_count["valid"] == False]
df_invalid
```

```python
df_validate_review_count["difference"] = (
    df_validate_review_count["review_count"] -
    df_validate_review_count["actual_review_count"]
)

df_validate_review_count["difference"].describe()
```

```python
df_validate_review_false = df_validate_review_count[df_validate_review_count["valid"] == False]
df_validate_review_false
```
